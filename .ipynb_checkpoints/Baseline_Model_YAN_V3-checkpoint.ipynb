```python
from cdqa.utils.converters import pdf_converter
from cdqa.utils.download import download_squad, download_model, download_bnpp_data
import pandas as pd
from ast import literal_eval
from cdqa.pipeline import QAPipeline
```

```python
df = pdf_converter(directory_path='./uploaded_disclosure_doc/')
df
```

```python
directory = './models/'

# Downloading data
download_squad(dir=directory)
download_bnpp_data(dir=directory)

# Downloading pre-trained BERT fine-tuned on SQuAD 1.1
download_model('bert-squad_1.1', dir=directory)

cdqa_pipeline = QAPipeline(reader='./models/bert_qa.joblib')
cdqa_pipeline.fit_retriever(df=df)
```

```python
def answer_question(question, number, score_weight):
    answer = cdqa_pipeline.predict(query=question, n_predictions=number, retriever_score_weight=score_weight)[0]
    answer_text = (
          "Answer:", answer[0], 
          "Answer Source Document:", answer[1],
          "Answer Context:", answer[2])
    return answer_text
```

```python
answer_question(simple_questions[0], N, weight)
```

```python
for q in simple_questions:
    lines_of_answer = answer_question(q, N, weight)
    print("Question:")
    print(q)
    for line in lines_of_answer:
        print(line)
    print()
```
